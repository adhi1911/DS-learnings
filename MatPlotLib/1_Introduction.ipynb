# Pieces of a matplotlib Graph


```python
x = np.array([-3,-2,-1,0,2,3])
y = x**2
x,y
```

```python
fig, ax = plt.subplots()

ax.plot(x,y)

plt.show()
```

```python
fig , axs = plt.subplots(1,2)
axs[0].plot(x,y)
axs[1].plot(x,x**3,color='green')

plt.show()

#subplots can go in multidimensions, also similar axs[] to plot different plots 
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y , color = 'red')
plt.show()
```

```python
a = np.arange(-5,5,0.01)
b = a**2
```

```python
fig , ax = plt.subplots(1,2)
ax[0].scatter(a,b)
ax[1].plot(a,b)
plt.show()
```

```python
b_damped_oscillation = np.exp(-a) * np.cos(2 * np.pi * a)

fig, ax = plt.subplots()
ax.plot(a,b_damped_oscillation , color = 'yellow')
plt.show()
```
