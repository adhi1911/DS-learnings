```python
import pandas as pd 
import numpy as np
import seaborn as sn 
```

```python
df = pd.read_csv('bhp.csv')
#column to be operated = price_per_sqft
df
```

```python
#draw the graph for price-per-sqft
sn.distplot(df['price_per_sqft'] , bins=20, kde=True)
```

```python
df['price_per_sqft'].describe().round(3)
#rounding off to 3 decimal places
```

```python
#1
#using percentile method to remove outliers
#we are removing the data which is not in the range of 0.001 to 0.999
min_threshold, max_threshold = df.price_per_sqft.quantile([0.001,0.999])
df[(df.price_per_sqft<min_threshold) | (df.price_per_sqft>max_threshold)]

#therefore 28 rows are outliers
```

```python
df_1 = df[(df.price_per_sqft>min_threshold) & (df.price_per_sqft<max_threshold)]
df_1.shape
```

```python
#plotting df_1
sn.distplot(df_1['price_per_sqft'] , bins=20, kde=True)
```

```python
#2 
#removing outliers using standard deviation
#we will use 4 standard deviation
max_limit = df_1.price_per_sqft.mean() + 4*df_1.price_per_sqft.std()
min_limit = df_1.price_per_sqft.mean() - 4*df_1.price_per_sqft.std()
max_limit, min_limit
df_1[(df_1.price_per_sqft>max_limit) | (df_1.price_per_sqft<min_limit)]
```

```python
df_2 = df_1[(df_1.price_per_sqft<max_limit) & (df_1.price_per_sqft>min_limit)]
df_2.shape
```

```python
#plotting df_2
sn.distplot(df_2['price_per_sqft'] , bins=20, kde=True)
```

```python
#3
#using 4 z score to remove outliers
#z = (x - mean)/std
df_2['zscore'] = (df_2.price_per_sqft-df_2.price_per_sqft.mean())/df_2.price_per_sqft.std()
df_2.sample(5)
```

```python
outliers_z = df_2[(df_2.zscore<-4) | (df_2.zscore>4)]
outliers_z
```
