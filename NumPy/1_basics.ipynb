## arrange

## reshape

```python
r1 = np.arange(0,15).reshape(3,5)
print(r1)
```

```python
#3d Tensor
r2 = np.arange(8 , dtype = int).reshape(2,2,2)
print(r2)
```

## ones and zeroes

default datatype is float

```python
o = np.ones((3,4),dtype = int)
print(o,type(o))
```

```python
z = np.zeros((4,5),dtype = complex)
print(z,type(z))
```

## Identity Matrix

```python
i3 = np.identity(3)
print(i3,type(i3))
```

```python
i5 = np.identity(5,dtype = int)
print(i5)
```
