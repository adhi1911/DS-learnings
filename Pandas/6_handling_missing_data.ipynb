```python
import pandas as pd 
import numpy as np

df = pd.read_csv('weather_data.csv')
df
```

```python
df2 = df.replace(np.nan, 'NA')
df2
```

```python
#replacing values in a column
# replace({column_name: {old_value: new_value}})
df3 = df2.replace({'temperature': {'NA','NaN'}}, 0)
df3


# we can pass in old value as a list or tuple or set or dictionary
# {'NA','NaN'} or ['NA','NaN'] or ('NA','NaN') or {'NA':'NaN'}

```

```python
#regex replace
#replace all the degree celcius with fahrenheit

df4 = df3.replace('[A-Za-z]', '', regex=True)
df4

#see it wiped out entire event column so we need to do it column wise
```

```python
df5 = df3.replace({
                    'temperature': '[A-Za-z]',
                    'windspeed': '[A-Za-z]'
                   }, '', regex=True)
#replacing empty string with 0
df5 = df5.replace('', 0)
df5
```
