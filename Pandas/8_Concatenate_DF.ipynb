```python
import pandas as pd

india_weather = pd.DataFrame({
    'city': ['mumbai', 'delhi', 'bangalore'],
    'temperature': [32, 45, 30],
    'humidity': [80, 60, 78]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    'city': ['new york', 'chicago', 'orlando'],
    'temperature': [21, 14, 35],
    'humidity': [68, 65, 75]
})

us_weather
```

```python
#concatenatin rows 
#here by default axis = 0
df = pd.concat([india_weather, us_weather])
df
```

```python

df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
#retrieving data for cities in india
#ignore_index will not work with this
df = pd.concat([india_weather, us_weather], keys=['India', 'USA'])
df
```

```python
df.loc['India']
```

```python
temperature_df = pd.DataFrame({
    'city': ['mumbai', 'delhi', 'bangalore'],
    'temperature': [32, 45, 30],
}, index=[0, 1, 2])

windspeed_df = pd.DataFrame({
    'city' :['mumbai', 'delhi', 'bangalore'],
    'windspeed': [7, 12, 9],
}, index=[0, 1, 2])


#we use index arguments to make sure that the index is same for both the dataframes
```

```python
#concatenating columns
df2 = pd.concat([temperature_df, windspeed_df], axis=1)
df2
```

```python
event_series = pd.Series(['Rain', 'Dry', 'Humid'], name='event')
event_series
```

```python
df3 = pd.concat([df2, event_series], axis=1)
df3
```
