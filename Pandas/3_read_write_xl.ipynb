```python
import pandas as pd 

#case: csv file has one extra header
#you want to skip that header
#you can use skiprows=1 or skiprows=[1] or skiprows=[0,1] or header = 1
df = pd.read_csv('stock_data.csv')
df
```

```python
df = pd.read_csv('stock_data.csv', skiprows=1)
df
```

```python
#what if no header?
#you can use header = None
df = pd.read_csv('stock_data.csv', header=None)
df
```

```python
#giving names to the columns
df = pd.read_csv('stock_data.csv', header = None, names=['ticker','eps','revenue','price','people'])
df
```

```python
#only read certain rows in dataframe
#use nrows
df = pd.read_csv('stock_data.csv', nrows=3)
#na values are not read
df = pd.read_csv('stock_data.csv', na_values=['not available','n.a.'])
df
```

```python
#writing to csv
#use to_csv

#write certain columns only
df.to_csv('new.csv')
```
