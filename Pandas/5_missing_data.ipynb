```python
df = pd.read_csv('5/weather_data.csv', parse_dates=['day'])
type(df.day[0])
```

```python
new_df = df.fillna(0)
new_df
```

```python
df2= df.fillna(method='ffill')
df2
#for na values it will take the previous value and fill it
#smiliar is bfill
```

```python
df3=df.fillna(method='bfill', axis='columns')
df3
#it will take the value from the next column and fill it
```

```python
df4=df.fillna(method='bfill', axis='rows')

#rows is default
df4
```

```python
#if there are consecutive na values we can use limit
df5=df.fillna(method='ffill', limit=1)
df5
#observer 4th adn 5th row
#5,6,7 were NA only 5 was filled
```
