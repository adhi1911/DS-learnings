```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart_disease_uci.csv')
df.head()
```

```python
df.dtypes
```

Is there any relation between age and cholestrol?


```python
df.groupby("age")[['chol']].mean().index   
```

```python
df.groupby("age")[['chol']].mean().values[:20]
```

Scatter Plot


```python
fig , ax = plt.subplots(1,2)
ax[0].scatter(df['age'].values , df['chol'].values, s=5)
ax[1].scatter(df.groupby("age")[['chol']].mean().index,df.groupby("age")[['chol']].mean().values , s = 12, c= 'red')
plt.show()
```

Line Plot


```python
fig , ax = plt.subplots()
ax.plot(df.groupby("age")[['chol']].mean().index,df.groupby("age")[['chol']].mean().values , c= 'violet')
plt.show()
```

Bar Graph

```python
fig , ax = plt.subplots()
ax.bar(df.groupby("age")[['chol']].mean().index,df.groupby("age")[['chol']].mean().values.ravel() , color= 'yellow')
plt.show()
```

Horizontal Bar Graph


```python
fig , ax = plt.subplots()
ax.barh(df.groupby("age")[['chol']].mean().index,df.groupby("age")[['chol']].mean().values.ravel() )
plt.show()
```
